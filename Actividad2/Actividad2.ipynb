**Ejercicio 1.** Se proporciona el siguiente programa para calcular el área de un rectángulo:

```
# Se utiliza el símbolo # para hacer comentarios.
# Se define una función de 2 variables para calcular el área.  
def area(x, y):
    return x*y

print("Proporciona el largo y altura del rectángulo: ", end="")
l = float(input())
w = float(input())
a = area(l, w)
# Imprime el valor del área en formato libre.
print("\nArea = ", a)
# Imprime el valor del área con 2 decimales. 
print("\nArea = {:.2f}".format(a))
```

En el programa que se nos proporciona podemos apreciar que primero se define la función "area" la cual depende de las variables $x$ y $y$, después se le pide al usuario ingresar los valores correspondientes para el largo y la altura del rectángulo, luego, el programa llama a la función definida previamente para evaluarla con los valores que otorga el usuario para después devolvernos el área calculada.

Realizaremos lo mismo para obtener el área de una elipse utilizando la fórmula $A_{elipse} = \pi a b$, donde $a$ es el semieje mayor y $b$ el semieje menor.

```
# Se importa la biblioteca NumPy
import numpy as np
# Se define una función de una variable para calcular el área de una elipse.  
def area(x,y):
    return np.pi*x*y

print("Proporciona el semieje mayor y el semieje menor de la elipse, respectivamente: ", end="")
a = float(input())
b = float(input())
c = area(a,b)
# Imprime el valor del área en formato libre.
print("\nArea = ", c)
# Imprime el valor del área con 2 decimales. 
print("\nArea = {:.2f}".format(c))
```

Ahora obtendremos el volumen de una esfera utilizando la fórmula $V_{esfera} = \frac{4}{3}\pi r^3$, donde $r$ es el radio de la esfera.

```
# Se importa la biblioteca NumPy
import numpy as np
# Se define una función de una variable para calcular el área de una elipse.  
def area(x):
    return (4/3)*np.pi*x**3

print("Proporciona el radio de la esfera: ", end="")
r = float(input())
a = area(r)
# Imprime el valor del área en formato libre.
print("\nArea = ", a)
# Imprime el valor del área con 2 decimales. 
print("\nArea = {:.2f}".format(a))
```

**Ejercicio 3.** Implementar el método Babilonio (o Método de Herón) para calcular la raíz cuadrada de un número S, cuando la sucesión converja a un error menor a $0.01$. Compare con el valor obtenido por el método Babilonio con el valor de la función **np.sqrt(S)**.

```
#Se importa la biblioteca NumPy
import numpy as np

#Se define la función que aproxima el error
def error(x):
    return (x**2-2)/(2*x)

print("Proporcione el número S del cual se quiere calcular la raíz cuadrada:")
S = float(input())
e = 1.0
Sreal = np.sqrt(S)

while e>=0.01:
    e = error(S)
    S -= e
print("La raíz cuadrada del número S con un error menor a 0.01 es:", S)
print("\nLa raíz cuadrada exacta del número S es:", Sreal)
```

```
#Importamos la biblioteca Matplotlib y NumPy
import matplotlib.pyplot as plt
import numpy as np

# Gráfica de funciones
# Definición de variables y funciones

x = np.linspace(-1.5,1.5,100)  # Arreglo de 100 puntos entre -1.5 y 1.5

y = np.log(abs(1+x))      # y será también un arreglo de 100 puntos automáticamente 


y4 = x - (1/2)*x**2 + (1/3)*x**3 - (1/4)*x**4

y7 = y4 + (1/5)*x**5 - (1/6)*x**6 + (1/7)*x**7

y11 = y7 - (1/8)*x**8 + (1/9)*x**9 - (1/10)*x**(10) + (1/11)*x**11

y16  = y11 - (1/12)*x**12 + (1/13)*x**13 - (1/14)*x**14 + (1/15)*x**15 - (1/16)*x**16


plt.subplot(111)

plt.xlabel("X label")

plt.ylabel("Y label")

plt.title("Gráfica de la actividad 4")

# Graficar varias funciones

plt.plot(x,y, label="ln(1+x)")

plt.plot(x,y4, label='T4')

plt.plot(x,y7, label='T7')

plt.plot(x,y11, label='T11')

plt.plot(x,y16, label='T16')

# Agregar una malla

plt.grid(True)

# Fijar los límites en el eje Y.

plt.ylim(-4, 2)

# Ubicación de las etiquetas de las funciones

plt.legend(loc="upper left")


plt.show()

#plt.savefig('plot.png')  # Si se desea guardar en un archivo la imagen de la gráfica
```
